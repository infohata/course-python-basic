## Teksto formatavimas

```python
vardas = "Kęstutis"
amzius = 40
tekstas = "Sveiki, aš " + \
    "esu " + vardas + \
    ", man yra " + str(amzius) + ' metų'
print(tekstas)
```

```python
tekstas = "Sveiki, aš esu %s ir man yra %d metų" \
    % (vardas, amzius)
print(tekstas)
```

```python
tekstas = "sveiki, aš esu {}, man yra {} metų".format(
    vardas, amzius
)
print(tekstas)
```

```python
tekstas = f"sveiki, aš esu {vardas}, man yra {amzius} metų"
print(tekstas)
```

```python
kaina = 99.9455
print(f"kaina: {kaina:.2f}")
```

```python
print(f"kaina: {kaina:010.0f}")
```

```python
print("Obuolys", "Kriaušė", "Bananas", sep=", ", end="; ")
print("Obuolys", "Kriaušė", "Bananas", sep=", ")
```
